###### 定义逻辑回归函数sigmoid

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
theta = np.matrix(theta)
theta = theta.T
theta.shape
```

###### 计算cost

```python
def compute_cost(X,theta,y):
    X = np.matrix(X)
    y = np.matrix(y)
    m = X.shape[0]#数据总行数
    s = sigmoid(np.dot(X,theta))
    # + 1e-6是为了避免计算log后的数值太大了，导致数据溢出，限制一下精度
    parm1 = np.multiply(-y,np.log(s + 1e-6))
    parm2 = np.multiply((1-y),np.log(1-s + 1e-6))
    return np.sum(parm1 - parm2) / m
```

```python
cost = compute_cost(X,theta,y)
cost
```

#### 下面进行梯度下降

采用向量化计算 𝑋.𝑇*(𝑆𝑖𝑔𝑚𝑜𝑖𝑑(𝑋𝜃)−𝑦)/m
X.T shape为（3，100），误差矩阵shape为（100，1）

```python
def compute_gradient(X,y,theta):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    m = X.shape[0]
    gradients = np.dot(X.T,(sigmoid(np.dot(X,theta)) - y))/m
    return gradients
```

```python
gradients = compute_gradient(X,y,theta)
gradients
```

```python
def update_parm(theta,gradients,alpha):
    theta = theta - np.dot(alpha,gradients)
    return theta
```

###### 进行批量梯度下降

```python
def gradientDescent(X,y,theta,alpha,iters):
    cost_history = []
    theta_history = []
    theta_history.append(theta)
    for i in range(iters):
        cost = compute_cost(X,theta,y)
        gradients = compute_gradient(X,y,theta)
        theta = update_parm(theta,gradients,alpha)
        cost_history.append(cost)
        theta_history.append(theta)
    return cost_history, theta_history, theta
```

```python
cost_history, theta_history, theta = gradientDescent(X,y,theta,0.0001,1000)
```

###### 编写一个函数进行预测，并计算准确率

当 ℎ𝜃
 大于等于0.5时，预测 y=1

当 ℎ𝜃
 小于0.5时，预测 y=0 。

```python
def predict(theta, X):
    probability = sigmoid(np.dot(X,theta))
    return [1 if x >= 0.5 else 0 for x in probability]
```

```python
predictions = predict(theta, X)
correct = [1 if ((a == 1 and b == 1) or (a == 0 and b == 0)) else 0 for (a, b) in zip(predictions, y)]
accuracy = (sum(map(int, correct)) % len(correct))
print ('accuracy = {0}%'.format(accuracy))
```
