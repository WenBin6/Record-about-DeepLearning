```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path =  'ex1data1.txt'
data = pd.read_csv(path, header=None, names=['Population', 'Profit'])
data.head()
```

```python
data.insert(0, 'Ones', 1)#添加一列方便向量运算
```

```python
data.head()#观察前五行
```

```python
#从数据集中提取出X和y
cols = data.shape[1]
X = data.iloc[:,0:cols-1]#去掉最后一列得到X
y = data.iloc[:,cols-1:cols]#保留最后一列得到y
```

```python
theta = np.random.rand(2,1)#随机生成theta，即w和b
```

```python
theta
```

```python
#X，y转为矩阵
X = np.matrix(X.values)
y = np.matrix(y.values)
```

```python
#计算线性模型预测出来的y
def compute_predict_y(X,theta):
    predict_y = np.dot(X,theta)
    return predict_y
```

```python
predict_y = compute_predict_y(X,theta)
predict_y
```

```python
#计算cost
def computeCost(y,predict_y):
    m = y.shape[0]
    cost = np.sum(np.power((y-predict_y),2))/(2*m)
    return cost
```

```python
print(computeCost(y,predict_y))
```

```python
#计算梯度,这个是比较难的点
def compute_gradient(X,y,predict_y):
    m = X.shape[0]
    #计算两个参数的梯度,矩阵乘法!!
    #X.T为2行m列，第一行代表自变量x1，第二行是常数1
    gradients = np.random.rand(2,1)
    gradients = np.dot(X.T,(predict_y - y))/m
    return gradients
```

```python
print(compute_gradient(X,y,predict_y))
```

```python
def update_parm(theta,gradients,alpha):
    #更新参数
    theta = theta - np.dot(alpha,gradients)
    return theta
```

```python
#迭代梯度下降
def gradientDescent(X,y,theta,alpha,iters):
    cost_history = []
    theta_history = []
    theta_history.append(theta)
    for i in range(iters):
        predict_y = compute_predict_y(X,theta)
        cost = computeCost(y,predict_y)
        gradients = compute_gradient(X,y,predict_y)
        theta = update_parm(theta,gradients,alpha)
        cost_history.append(cost)
        theta_history.append(theta)
    return cost_history, theta_history, theta
```

```python
cost_history, theta_history, theta = gradientDescent(X,y,theta,0.001,1000)
```
